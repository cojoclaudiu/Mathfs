Scrieţi un program care citeste de la utilizator numere pana cand acesta introduce “gata”. Dupa introducerea
acestui cuvant vor fi afisate suma numerelor, numarul de valori introduse si media lor. Daca utilizatorul introduce
altceva decat un numar, acest lucru va fi detectat prin utilizarea try/except si se va afisa un mesaj de eroare, iar
programul va solicita introducerea unui nou numar, fara a opri executia.

```python
def citeste_numere():
  # Suma totala a calculului
  suma_totala = 0

  # Numarul total de valori introduse de utilizator
  numar_valori = 0

  while True:
    valoare = input("Introduceti un numar (sau 'gata' pentru a termina calculul.)")

    if valoare.lower() == "gata":
      break

    try:
      numar = float(valoare)
      suma_totala += numar
      numar_valori += 1

    except ValueError:
      print("Eroare: trebuie sa introduceti un numar valid!")

  if numar_valori > 0:
    media = suma_totala / numar_valori
    print(f"\nSuma numerelor: {suma_totala}")
    print(f"Numarul de valori introduse: {numar_valori}")
    print(f"Media valorilor: {media}")
  else:
    print("\nNu a fost introdus niciun numar.")

if __name__ == "__main__":
  citeste_numere()
```
