Scrieti un program care solicită utilizatorului introducerea de la tastatură a numărului de ore
de lucru şi rata orară, calculând apoi salariul brut.

```python
def calcul_salariu_brut():
  salariu_brut = 0

  while True:
    try:
      numar_ore_de_lucru = float(input("Introduceti numarul de ore lucrate: "))
      rata_orara = float(input("Introduceti rata orara: "))
      salariu_brut = numar_ore_de_lucru * rata_orara
      print(f"Salariul brut este: {salariu_brut:.2f}: lei")
      break

    except ValueError:
      print("Eroare: trebuie sa introduceti valori numerice!\nIncearca din nou.\n")

if __name__ == "__main__":
  calcul_salariu_brut()
```
