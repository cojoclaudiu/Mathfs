Scrieti un program care solicită utilizatorului introducerea de la tastatură a numărului de ore
de lucru şi rata orară, calculând apoi salariul brut.

```python
def calcul_salariu_brut():
  salariu_brut = 0

  while True:
    try:
      numar_ore_de_lucru = float(input("Introduceti numarul de ore lucrate: "))
      rata_orara = float(input("Introduceti rata orara: "))
      salariu_brut = numar_ore_de_lucru * rata_orara
      print(f"Salariul brut este: {salariu_brut:.2f}: lei")
      break

    except ValueError:
      print("Eroare: trebuie sa introduceti valori numerice!\nIncearca din nou.\n")

if __name__ == "__main__":
  calcul_salariu_brut()
```

Scrieti un program care solicita introducerea temperaturii în grade Celsius, o converteste în
Fahrenheit si afiseaza valoarea transformata.

```python
def transforma_celsius_in_fahrenheit():
  fahrenheit = 0

  while True:
    try:
      celsius = float(input("Introduceti numarul de grade Celsius pentru a le transforma in Fahrenheit: "))
      fahrenheit = celsius * 9 / 5 + 32
      print(f"{celsius} de grade Celsius in Fahrenheit sunt egale cu {fahrenheit}")
      break

    except ValueError:
       print("Eroare: trebuie sa introduceti valori numerice!\nIncearca din nou.\n")

if __name__ == "__main__":
  transforma_celsius_in_fahrenheit()
```
